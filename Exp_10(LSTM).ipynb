```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

seq_length = int(input("Enter sequence length: "))
data_points = int(input("Enter number of data points: "))

print(f"Enter {data_points} values:")
data = []
for i in range(data_points):
    data.append(float(input()))

data = np.array(data).reshape(-1, 1)
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)

X, y = [], []
for i in range(len(data_scaled) - seq_length):
    X.append(data_scaled[i:i+seq_length])
    y.append(data_scaled[i+seq_length])

X = np.array(X)
y = np.array(y)

model = Sequential([
    LSTM(50, activation='relu', input_shape=(seq_length, 1)),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
epochs = int(input("Enter number of epochs: "))
model.fit(X, y, epochs=epochs, verbose=0)

print("\nPredictions:")
predictions = model.predict(X, verbose=0)
predictions = scaler.inverse_transform(predictions)

for i, pred in enumerate(predictions):
    print(f"Step {i+1}: {pred[0]:.2f}")

last_seq = data_scaled[-seq_length:].reshape(1, seq_length, 1)
next_pred = model.predict(last_seq, verbose=0)
next_pred = scaler.inverse_transform(next_pred)
print(f"\nNext predicted value: {next_pred[0][0]:.2f}")
```
